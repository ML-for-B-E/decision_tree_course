# Question 1 : Nombre moyen de questions posées par partie

Pour déterminer le nombre moyen de questions posées, on calcule l'espérance du nombre de questions nécessaires pour identifier un personnage.

## Formule générale

L'espérance est donnée par :

$$
E(Q)=\sum_{i=1}^{n} P(i)\times d_i
$$

où :

- $P(i)$ est la probabilité de sélectionner le personnage $i$ ;
- $d_i$ est le nombre de questions nécessaires pour identifier ce personnage.

Comme les 4 personnages sont supposés équiprobables :

$$
P(i)=\frac{1}{4}
$$


## Arbre de décision (a)

| Personnage | Nombre de questions |
|------------|--------------------|
| Brian | 1 |
| John | 2 |
| Aphra | 3 |
| Aoife | 3 |

L'espérance vaut donc :

$$
E_a=\frac{1}{4}(1+2+3+3)
$$

$$
E_a=\frac{9}{4}
$$

$$
E_a=2.25
$$


## Arbre de décision (b)

| Personnage | Nombre de questions |
|------------|--------------------|
| Brian | 1 |
| Aphra | 2 |
| John | 3 |
| Aoife | 3 |

L'espérance vaut :

$$
E_b=\frac{1}{4}(1+2+3+3)
$$

$$
E_b=\frac{9}{4}
$$

$$
E_b=2.25
$$


## Conclusion

Le nombre moyen de questions posées est identique pour les deux arbres :

$$
E_a = E_b = 2.25
$$

Ainsi, en moyenne, il faut **2,25 questions** pour identifier un personnage, quel que soit l'arbre de décision utilisé.

# Question 2 : Nombre moyen de questions posées par partie

Dans cet arbre de décision, chaque personnage est identifié après exactement deux questions.

| Personnage | Nombre de questions |
|------------|--------------------|
| Brian | 2 |
| John | 2 |
| Aphra | 2 |
| Aoife | 2 |

Comme les quatre personnages sont supposés équiprobables, la probabilité de sélectionner chaque personnage est :

$$
P(i)=\frac{1}{4}
$$

L'espérance du nombre de questions est donnée par :

$$
E(Q)=\sum_{i=1}^{n} P(i)\times d_i
$$

où :

- $P(i)$ est la probabilité du personnage $i$ ;
- $d_i$ est le nombre de questions nécessaires pour l'identifier.

En remplaçant par les valeurs observées :

$$
E(Q)=\frac{1}{4}(2+2+2+2)
$$

$$
E(Q)=\frac{8}{4}
$$

$$
E(Q)=2
$$

## Conclusion

Le nombre moyen de questions posées par partie est :

$$
\boxed{2}
$$

Ainsi, cette nouvelle série de questions permet d'identifier un personnage en **2 questions en moyenne**, soit une amélioration par rapport aux arbres précédents dont la moyenne était de **2,25 questions**.
