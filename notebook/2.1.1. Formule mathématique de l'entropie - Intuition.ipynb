## Intuition derrière la formule de l'entropie

L'entropie mesure le niveau d'incertitude ou d'impureté d'un ensemble de données.

L'idée est la suivante :

- Si tous les éléments appartiennent à la même catégorie, il n'y a aucune incertitude.
- Si les éléments sont répartis équitablement entre plusieurs catégories, l'incertitude est maximale.

La formule de l'entropie est :

$$
H(t) = - \sum_{i=1}^{l} P(t=i)\log_2(P(t=i))
$$

Chaque terme :

$$
-P(t=i)\log_2(P(t=i))
$$

représente la quantité d'information associée à la catégorie $i$.


### Cas 1 : Ensemble parfaitement homogène

Supposons que tous les individus appartiennent à une seule catégorie :

$$
P(A)=1
$$

L'entropie vaut alors :

$$
H = -(1)\log_2(1)
$$

Or :

$$
\log_2(1)=0
$$

Donc :

$$
H=0
$$

Il n'y a aucune incertitude puisque la catégorie est connue à l'avance.


### Cas 2 : Deux catégories équiprobables

Supposons :

$$
P(A)=0.5
$$

$$
P(B)=0.5
$$

L'entropie devient :

$$
H=-(0.5\log_2(0.5)+0.5\log_2(0.5))
$$

Comme :

$$
\log_2(0.5)=-1
$$

on obtient :

$$
H=-(0.5(-1)+0.5(-1))
$$

$$
H=1
$$

L'incertitude est maximale car les deux catégories sont également probables.


### Interprétation dans un arbre de décision

L'objectif d'un arbre de décision est de choisir les questions qui réduisent le plus rapidement l'entropie.

Ainsi :

- Une entropie élevée signifie qu'il reste beaucoup d'incertitude.
- Une entropie faible signifie que les données sont plus homogènes.
- Une entropie nulle signifie que la réponse est connue avec certitude.

Une bonne question est donc celle qui diminue fortement l'entropie après la séparation des données.


L'entropie mesure le degré d'incertitude d'un ensemble de données. Plus l'entropie est faible, plus les données sont homogènes et plus il est facile de prendre une décision.
