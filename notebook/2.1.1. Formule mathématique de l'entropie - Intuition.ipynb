## Intuition derrière la formule de l'entropie

### Un exemple pour se mettre en jambes

Imaginons un médecin qui analyse deux groupes de patients :

| Groupe | Malades | Sains |
|--------|---------|-------|
| A      | 8       | 0     |
| B      | 4       | 4     |

- **Groupe A** : tous malades. Le diagnostic est évident — aucune incertitude.
- **Groupe B** : moitié malades, moitié sains. Impossible de prédire sans information supplémentaire — incertitude maximale.

L'entropie est la mesure qui **quantifie cette incertitude**. Plus l'entropie est élevée, plus les données sont mélangées et difficiles à classer.

### Pourquoi un logarithme ?

Avant d'écrire la formule, il faut comprendre ce que représente $-\log_2(p)$.

Considérons un événement de probabilité $p$ :

- Si $p = 1$ (événement certain), apprendre qu'il s'est produit n'apporte **aucune information** — on le savait déjà.
- Si $p = 0.01$ (événement très rare), apprendre qu'il s'est produit est **très surprenant** — cela apporte beaucoup d'information.

On définit la **quantité d'information** associée à un événement par :

$$
I(p) = -\log_2(p)
$$

| $p$ | $I(p) = -\log_2(p)$ | Interprétation |
|-----|----------------------|----------------|
| 1    | 0 bit   | Certain — aucune surprise |
| 0.5  | 1 bit   | Équiprobable (pile ou face) |
| 0.25 | 2 bits  | Assez rare |
| 0.01 | ≈ 6.6 bits | Très rare — très surprenant |

> La base 2 donne une unité naturelle : le **bit**. Un résultat de pile ou face apporte exactement 1 bit d'information.

L'entropie est simplement la **valeur moyenne** de cette surprise sur toutes les catégories.

### La formule de l'entropie

$$
H = - \sum_{i=1}^{k} p_i \log_2(p_i)
$$

où :
- $k$ est le nombre de catégories,
- $p_i$ est la proportion d'éléments appartenant à la catégorie $i$.

> **Note :** Dans certains cours de Machine Learning, cette formule s'écrit $H(t) = -\sum P(t=i)\log_2 P(t=i)$, ce qui est équivalent avec $p_i = P(t=i)$.

> **Convention mathématique :** Par continuité, on pose $0 \cdot \log_2(0) = 0$. Une catégorie absente ne contribue pas à l'entropie.

### Cas 1 : Ensemble parfaitement homogène

Reprenons le **Groupe A** (8 malades, 0 sains) :

$$p_{\text{malade}} = 1, \quad p_{\text{sain}} = 0$$

$$H = -\bigl(1 \cdot \log_2(1) + 0 \cdot \log_2(0)\bigr)$$

Avec $\log_2(1) = 0$ et la convention $0 \cdot \log_2(0) = 0$ :

$$\boxed{H = 0}$$

Il n'y a aucune incertitude — la catégorie est connue à l'avance.

### Cas 2 : Deux catégories équiprobables

Reprenons le **Groupe B** (4 malades, 4 sains) :

$$p_{\text{malade}} = 0{,}5, \quad p_{\text{sain}} = 0{,}5$$

$$H = -\bigl(0{,}5 \cdot \log_2(0{,}5) + 0{,}5 \cdot \log_2(0{,}5)\bigr)$$

Avec $\log_2(0{,}5) = -1$ :

$$H = -\bigl(0{,}5 \cdot (-1) + 0{,}5 \cdot (-1)\bigr)$$

$$\boxed{H = 1 \text{ bit}}$$

L'incertitude est maximale : les deux catégories sont également probables.

```python
import numpy as np
import matplotlib.pyplot as plt

# Entropie binaire
p = np.linspace(1e-9, 1 - 1e-9, 500)
H = -(p * np.log2(p) + (1 - p) * np.log2(1 - p))

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(p, H, color='steelblue', linewidth=2.5, label='H(p)')

# Points remarquables
ax.scatter([0, 0.5, 1], [0, 1, 0], color='tomato', zorder=5, s=70)
ax.annotate("H = 0\n(groupe homogène)", xy=(0, 0), xytext=(0.07, 0.28),
            arrowprops=dict(arrowstyle='->', color='gray'), fontsize=10)
ax.annotate("H = 1 bit\n(incertitude max)", xy=(0.5, 1), xytext=(0.6, 0.82),
            arrowprops=dict(arrowstyle='->', color='gray'), fontsize=10)
ax.annotate("H = 0\n(groupe homogène)", xy=(1, 0), xytext=(0.76, 0.28),
            arrowprops=dict(arrowstyle='->', color='gray'), fontsize=10)

ax.axvline(0.5, color='gray', linestyle='--', alpha=0.5, linewidth=1)
ax.set_xlabel("Proportion $p$ de la classe A", fontsize=12)
ax.set_ylabel("Entropie $H$ (bits)", fontsize=12)
ax.set_title("Entropie d'un ensemble binaire (2 classes)", fontsize=13)
ax.set_xlim(0, 1)
ax.set_ylim(-0.05, 1.15)
ax.legend(fontsize=11)
plt.tight_layout()
plt.show()
```

### Cas 3 : Trois catégories équiprobables

Imaginons un groupe de 9 patients avec 3 diagnostics possibles :

| Diagnostic | Effectif | Proportion $p_i$ |
|------------|----------|------------------|
| Sain       | 3        | $1/3$            |
| Malade A   | 3        | $1/3$            |
| Malade B   | 3        | $1/3$            |

$$H = -\left(\frac{1}{3}\log_2\frac{1}{3} + \frac{1}{3}\log_2\frac{1}{3} + \frac{1}{3}\log_2\frac{1}{3}\right)$$

$$H = -\log_2\frac{1}{3} = \log_2 3 \approx 1{,}58 \text{ bits}$$

> **Généralisation :** Pour $k$ catégories équiprobables, l'entropie maximale vaut $\log_2(k)$.
> - 2 classes → max = 1 bit
> - 3 classes → max ≈ 1,58 bits
> - 4 classes → max = 2 bits

### Interprétation dans un arbre de décision

L'objectif d'un arbre de décision est de choisir les questions (critères de séparation) qui **réduisent le plus rapidement l'entropie**.

| Situation | Entropie | Signification |
|-----------|----------|---------------|
| Avant la question | Élevée | Beaucoup d'incertitude |
| Après une bonne question | Faible | Groupes plus homogènes |
| Nœud feuille idéal | Nulle | Catégorie connue avec certitude |

Une **bonne question** est donc celle qui crée des sous-groupes à entropie faible. C'est exactement ce que mesure le **gain d'information** — la différence d'entropie avant et après la séparation.
